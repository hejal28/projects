```python
#bit flip, the goal is to create |1> from |0>
def create_circuit():
    qc = QuantumCircuit(1)

    qc.x(0)

    return qc

# check solution
qc2 = create_circuit()
state = Statevector(qc2)
```

```python
plot_state_qsphere(state, show_state_labels=True, show_state_phases = True, use_degrees = True);
```

```python
#superposition to reach |+> state
def create_circuit2():
    qc = QuantumCircuit(1)
    qc.h(0)
    return qc

qc3 = create_circuit2()
state = Statevector(qc3)
plot_state_qsphere(state, show_state_labels=True, show_state_phases = True, use_degrees = True);
```
